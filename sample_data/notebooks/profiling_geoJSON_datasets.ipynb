# Profiling and Enriching GeoJSON Datasets

This code iterates over all GeoJSON files in the geo_datasets_sample folder, and for each file:

- Loads it into a GeoDataFrame using geopandas.
- Profiles the dataset attributes (columns, data types, distinct values, missing values, etc.) using datamart_profiler.
- Extracts top-level GeoJSON metadata (geometricType, name, crs) from the original file.
- Combines the profile and metadata into a single enriched dictionary.
- Saves the enriched metadata as a JSON file in the geo_datasets_metadata folder, with _metadata.json appended to the original filename.

This allows for easy documentation, exploration, and analysis of both the attribute data and spatial metadata for all GeoJSON datasets in the folder.

    

```python
import os
import json
import geopandas as gpd
import datamart_profiler

def profile_geojson(input_path, output_dir):
    """
    Reads a GeoJSON file, profiles it, enriches with GeoJSON metadata,
    and saves the profile as a JSON in output_dir.
    """
    # Load GeoJSON into GeoDataFrame
    gdf = gpd.read_file(input_path)
    
    # Profile the dataset
    profile = datamart_profiler.process_dataset(gdf, include_sample=True, plots=False)
    
    # Load raw GeoJSON to extract metadata
    with open(input_path, "r") as f:
        geojson_data = json.load(f)
    
    geojson_metadata = {
        "geometricType": geojson_data.get("geometricType"),
        "name": geojson_data.get("name"),
        "crs": geojson_data.get("crs")
    }
    
    # Merge metadata and profile
    enriched_profile = {**geojson_metadata, **profile}
    
    # Prepare output path
    os.makedirs(output_dir, exist_ok=True)
    base_name = os.path.basename(input_path).replace(".geojson", "_metadata.json")
    output_path = os.path.join(output_dir, base_name)
    
    # Save enriched profile as JSON
    with open(output_path, "w") as f:
        json.dump(enriched_profile, f, indent=2)
    
    print(f"Saved metadata for {input_path} -> {output_path}")
```

```python
input_dir = "../geojson" # sample geo datasets
output_dir = "../metadata" # store metadata

for fname in os.listdir(input_dir):
    if fname.endswith(".geojson"):
        path = os.path.join(input_dir, fname)
        profile_geojson(path, output_dir)
```
